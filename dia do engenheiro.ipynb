```python
# Importando bibliotecas necessárias
# Pillow para manipulação de imagens
from PIL import Image, ImageDraw, ImageFont
```

```python
# Vamos acompanhar a imagem ser construída no passo-a-passo, tudo bem? :D 
# +++ Calma Stênio, Vini e outros pupilos que ficam muito ansiosos... kkkkk +++
```

```python
# Criando uma imagem em branco
img = Image.new('RGB', (1080, 1080), color='white')
d = ImageDraw.Draw(img)
# Somente o canvas...
display(img)
```

```python
# Carregando fontes
fonte_titulo = ImageFont.truetype("arial.ttf", 60)
fonte_texto = ImageFont.truetype("arial.ttf", 40)
# Essa reservarei para a cereja do bolo ;)
fonte_texto2 = ImageFont.truetype("ROCKEB.TTF", 30)

# Adicionando texto à imagem
d.text((540, 200), "Feliz Dia do Engenheiro Químico!", fill=(0,0,0), font=fonte_titulo, anchor="mm", align="center")
# Texto se formando ;)
display(img)
```

```python
# Nossa data !
data = '20/09/2024'
d.text((540, 900), f"Celebrado em {data}", fill=(100,100,100), font=fonte_texto, anchor="mm")
# Texto final
display(img)
```

```python
# Agora vamos "adornar" (feat profa. Andréa kkk) a coisa ;)
# Desenhando um béquer estilizado - vamos tentar, né!!! kkkkk
d.rectangle([440, 400, 640, 800], outline=(0,0,0), width=5)
d.rectangle([490, 350, 590, 400], outline=(0,0,0), width=5)
d.line([440, 400, 490, 350], fill=(0,0,0), width=5)
d.line([640, 400, 590, 350], fill=(0,0,0), width=5)
# Vamos ver como ficou?
display(img)
```

```python
# Agora o toque final !
# Adicionando "líquido" colorido no béquer
for y in range(600, 800):
        cor = (int(255 * (y-600) / 200), 0, 255 - int(255 * (y-600) / 200))
        d.line([441, y, 639, y], fill=cor)
        
d.text((540, 1000), "Parabéns queridos colegas PROFESSORES e ALUNOS! \nAss. SamCartx", fill=(255,0,0), font=fonte_texto2, anchor="mm")                            
```

```python
# Salvando a imagem - DE LEMBRANÇA PARA NÓS !!!
img.save("dia_engenheiro_quimico.png")

# Vamos ver a figura completa?
display(img)
```
